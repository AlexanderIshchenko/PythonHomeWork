# f(x) = 0.6x^3+5.5x^2+10x -5


1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
# п.1 и п.4 Находим корни (округляем для удобства до сотых) и строим график
from sympy import *
from sympy.plotting import plot 
x = Symbol('x')
f = 0.6*x**3 + 5.5*x**2 + 10*x - 5
lst = [round(x,2) for x in solve(f,x)]
print(f'Корни функции: {lst}')
plot(f)
```

```python
# п.5 Находим вершины
lst_sum_x = solve(f.diff(x),x)
lst_sum_y = [(0.6*x**3 + 5.5*x**2 + 10*x - 5) for x in lst_sum_x]
res = list(zip(lst_sum_x, lst_sum_y))
print(f'Вершины функции: {res}')
```

```python
# п.2 и п.3
print(f'Функция возрастает на промежутке от (-00 до {lst_sum_x[0]}) и от ({lst_sum_x[1]} до +00)')
print(f'Функция убывает на промежутке от ({lst_sum_x[0]} до {lst_sum_x[1]})')
```
